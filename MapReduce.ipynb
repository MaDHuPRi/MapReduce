```python
def clean_word(word):
    """Removes punctuation except for valid apostrophes in contractions."""
    # Allow words like "I'm", "he's", "don't" while removing other punctuation
    cleaned = re.sub(r"[^\w\s']", "", word)  # Keeps letters, numbers, and apostrophes
    return cleaned.lower()
```

Question 1.
Write Python code and use MapReduce to count occurrences of each word in the first text file (file.txt). How many times each word is repeated?

```python
#common reducer and run_mapreduce functions for both the questions

def reducer(mapped_data):
    #Reduces the mapped data by summing occurrences of each word.
    word_counts = defaultdict(int)
    for word, count in mapped_data:
        word_counts[word] += count
    return word_counts

def run_mapreduce(file_path):
    #Runs MapReduce, and prints word counts
    mapped_data = []

    with open(file1, "r") as file:
        for line in file:
            mapped_data.extend(mapper(line))  # Mapper function

    word_counts = reducer(mapped_data)  # Reduce


    for word, count in sorted(word_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"{word}: {count}")
```

```python
def mapper(line):
    #Processes a line and emits a list of (word, 1) """
    mapped_results = []  # Ensure this is always a list

    words = line.strip().split()
    for word in words:
        word = clean_word(word)
        mapped_results.append((word, 1))

    return mapped_results  #returns a tuple (word,1) for every word in the line

# Running MapReduce on the file
run_mapreduce(file1)
```
